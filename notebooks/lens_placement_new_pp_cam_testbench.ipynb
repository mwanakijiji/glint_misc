```python
# Finds the position of f-length of a lens to image both the DM and a new PP cam
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def d_i(f, d_o):
    
    d_i_pass = np.divide(f*d_o, d_o-f)
    m_pass = -d_i_pass/d_o
    
    return d_i_pass, m_pass
```

```python
x_1 = np.linspace(1,100,100)
x_2 = np.linspace(140,200,100)
```

```python
d_i_channel_1, m_channel_1 = d_i(f=250, d_o=x_1-150)
d_i_channel_2, m_channel_2 = d_i(f=250, d_o=x_2)
```

```python
plt.scatter(x_1, d_i_channel_1, label='di, channel 1')
plt.scatter(x_2, d_i_channel_2, label='di, channel 2')
plt.xlabel('x1 or x2')
plt.ylabel('di')
plt.show()
```

```python
plt.scatter(x_1, m_channel_1)
plt.xlabel('M, channel 1')
plt.ylabel('M')
plt.show()
```

```python
plt.scatter(x_2, d_i_channel_2)
plt.ylabel('di, channel 2')
plt.xlabel('x2')
plt.show()
```

```python
plt.scatter(x_2, m_channel_2)
plt.xlabel('di, channel 2')
plt.ylabel('M')
plt.show()
```
