```python

# Read the JPEG image
image_path = stem + 'perrin_hr4796A.jpeg'
image = Image.open(image_path)

# Create a figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(image)

radius = 20

# Add a red circle
circle = patches.Circle((1200, 290), radius, edgecolor='red', facecolor='none')
ax.add_patch(circle)

# Show the plot
plt.show()
```
